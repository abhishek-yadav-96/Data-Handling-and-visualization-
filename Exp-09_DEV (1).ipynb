```python
#3049
from datascience import *
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
# Assuming "dog.csv" is in the same directory as your script or notebook
dogs = pd.read_csv("dogs.csv")
print(dogs)
```

```python
#3049
dogs["Height(cm)"].hist()
plt.show()
```

```python
#3049
dogs["Height(cm)"].hist(bins=20)
plt.show()
```

```python
#3049
dogs["Height(cm)"].hist(bins=5)
plt.show()
```

```python
#3049
avg_weight_by_breed = dogs.groupby("Breed")["Weight(kg)"].mean()
print(avg_weight_by_breed)
```

```python
#3049
avg_weight_by_breed.plot(kind="bar")
plt.show()
```

```python
#3049
avg_weight_by_breed.plot(kind="bar",
title="Mean Weight by Dog Breed")
plt.show()
```

```python
#3049
#Layering plots
dogs[dogs["Sex"]=="F"]["Height(cm)"].hist()
dogs[dogs["Sex"]=="M"]["Height(cm)"].hist()
plt.show()
```

```python
#3049
dogs[dogs["Sex"]=="F"]["Height(cm)"].hist(alpha=0.7)
dogs[dogs["Sex"]=="M"]["Height(cm)"].hist(alpha=0.7)
plt.legend(["F", "M"])
plt.show()
```
